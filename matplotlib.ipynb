```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.xlabel("X-axis")  
plt.ylabel("Y-axis")
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-', linewidth=2)
plt.grid(True)
plt.plot(x,y,linestyle='--', color='red', linewidth=2)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
plt.title("Simple Line Graph")
plt.xlabel("Months")  
plt.ylabel("Savings account")
plt.plot(x,y,marker='o',linestyle='--', color='green')
plt.show()
```

```python
c=['A','B','C','D','E']
v=[10,15,7,12,9]
plt.bar(c,v,color='b')
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

c = ['BMW', 'AUDI', 'BENZ', 'PORSCHE']
v = [30, 20, 10, 40]

colors=['pink','violet','lavender','purple']
plt.pie(v, labels=c,colors=colors)


plt.axis('equal')  

plt.show()
```

```python
c=['A','B','C','D','E']
v=[10,15,7,12,9]
plt.scatter(c,v,color='b')
plt.title("Basic Scatter Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
