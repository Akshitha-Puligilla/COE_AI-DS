```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.xlabel("X-axis")
plt.ylabel("y axis")
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-',linewidth=1.5)
plt.plot(x,y,linewidth=2,linestyle='--')
plt.show()
```
