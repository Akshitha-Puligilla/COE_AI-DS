```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("C:/Users/CVR/Downloads/train.csv")
```

```python
sns.barplot(data=df,x="Pclass",y="Survived",errorbar=None,palette='rocket')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('passenger')
plt.ylabel('survived')
plt.show()
```

```python
df
```

```python

sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(df[['Fare', 'Age', 'Pclass']],hue='Pclass')
plt.show()
```

```python
numeric_cols=df.select_dtypes(include='number')
corr_mat=numeric_cols.corr()
sns.heatmap(corr_mat,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title('correlation matrix heatmap')
plt.show()
```
